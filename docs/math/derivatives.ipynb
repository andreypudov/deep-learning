# Derivatives

Derivatives point in the direction of increases and decreases in a mathematical funciton.

In deep learning, the goal (e.g., classification) is represented as an error of function. This, the best solution is the point with the smallest error.

The derivative tells us which way to "move" in that error landscape in order to find the optimal solution.

## Derivative of polynomials

$$
\frac{d}{dx} (ax^n) = n \cdot a \cdot x^{n-1}
$$

$$
f(x) = x^2
$$

$$
f(x)' = 2x
$$

```python
import numpy as np
import matplotlib.pyplot as plt


# Define the function and its derivative
def f(x):
    return x**2


def df(x):
    return 2 * x


# Create a range of x values
x = np.linspace(-4, 4, 400)

# Compute y values
y = f(x)
dy = df(x)

# Create the figure
plt.figure(figsize=(12, 5))

# Plot f(x)
plt.subplot(1, 2, 1)
plt.plot(x, y, label=r"$f(x) = x^2$", color="cyan")
plt.axhline(0, color="white", linewidth=0.5)
plt.axvline(0, color="white", linewidth=0.5)
plt.title("Function: $x^2$")
plt.grid(True, linestyle="--", alpha=0.3)
plt.legend()

# Plot f'(x)
plt.subplot(1, 2, 2)
plt.plot(x, dy, label=r"$f'(x) = 2x$", color="orange")
plt.axhline(0, color="white", linewidth=0.5)
plt.axvline(0, color="white", linewidth=0.5)
plt.title("Derivative: $2x$")
plt.grid(True, linestyle="--", alpha=0.3)
plt.legend()

plt.tight_layout()
plt.show()
```

## Visual intuition

Imagine walking along the parabola.

At each step, you ask:

    “How steep is the ground right here?”

The derivative graph 2x answers that question.

- When you’re left of the vertex, the slope is negative → you’re walking downhill.
- At the vertex, slope is zero → flat.
- To the right, slope is positive → uphill.
- The farther you go, the steeper it gets → because 2x grows.

The straight line shape reflects the fact that the parabola’s slope increases at a constant rate.

## Minima and maxima

The derivative is zero at the vertex of the parabola, which is the minimum point. This is because the slope changes from negative to positive there. Similarly, the derivative is zero at the maximum point of a parabola, where the slope changes from positive to negative.

```python
import numpy as np
import matplotlib.pyplot as plt


# Define the function and its derivative
def f(x):
    return -(x**4) + 3 * x**2


def df(x):
    return -4 * x**3 + 6 * x


# Range of x values
x = np.linspace(-3, 3, 400)

# Compute values
y = f(x)
dy = df(x)

# Create the figure
plt.figure(figsize=(8, 5))

# Plot both on the same axes
plt.plot(x, y, label=r"$f(x) = -x^4 + 3x^2$", color="cyan")
plt.plot(x, dy, label=r"$f'(x) = -4x^3 + 6x$", color="orange")

# Axes lines
plt.axhline(0, color="white", linewidth=0.5)
plt.axvline(0, color="white", linewidth=0.5)

# Limit x-axis
plt.xlim(-2, 2)
plt.ylim(-3, 3)

plt.title("Function and Derivative on the Same Axes")
plt.grid(True, linestyle="--", alpha=0.3)
plt.legend()

plt.tight_layout()
plt.show()
```

$$
\begin{aligned}
4x^3 + 6x &= 0 \\
x(4x^2 + 6) &= 0 \\
x &= 0 \quad \text{or} \quad 4x^2 + 6 = 0 \\
x &= 0 \quad \text{or} \quad x^2 = -\frac{6}{4} \\
x &= 0 \quad \text{or} \quad x^2 = -\frac{3}{2} \\
x &= 0 \quad \text{or} \quad x = \pm \sqrt{\frac{3}{2}}
\end{aligned}
$$

- Minima and maxima have $df = 0$.
- Minima have $df < 0$ to the left and $df > 0$ to the right.
- Maxima have $df > 0$ to the left and $df < 0$ to the right.

```python
import numpy as np
import matplotlib.pyplot as plt


def f(x):
    return x * np.sin(x**2) + 1


x = np.linspace(-3, 3, 400)

y = f(x)

plt.figure(figsize=(8, 5))

plt.plot(x, y, label=r"$f(x) = xsin(x^2) + 1$", color="blue")

plt.axhline(0, color="white", linewidth=0.5)
plt.axvline(0, color="white", linewidth=0.5)

plt.xlim(-2, 2)
plt.ylim(-1, 3)

plt.grid(True, linestyle="--", alpha=0.3)
plt.legend()

plt.tight_layout()
plt.show()
```
