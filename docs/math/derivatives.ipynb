# Derivatives

Derivatives point in the direction of increases and decreases in a mathematical funciton.

In deep learning, the goal (e.g., classification) is represented as an error of function. This, the best solution is the point with the smallest error.

The derivative tells us which way to "move" in that error landscape in order to find the optimal solution.

## Derivative of polynomials

$$
\frac{d}{dx} (ax^n) = n \cdot a \cdot x^{n-1}
$$

$$
f(x) = x^2
$$

$$
f(x)' = 2x
$$

```python
import numpy as np
import matplotlib.pyplot as plt


# Define the function and its derivative
def f(x):
    return x**2


def df(x):
    return 2 * x


# Create a range of x values
x = np.linspace(-4, 4, 400)

# Compute y values
y = f(x)
dy = df(x)

# Create the figure
plt.figure(figsize=(12, 5))

# Plot f(x)
plt.subplot(1, 2, 1)
plt.plot(x, y, label=r"$f(x) = x^2$", color="cyan")
plt.axhline(0, color="white", linewidth=0.5)
plt.axvline(0, color="white", linewidth=0.5)
plt.title("Function: $x^2$")
plt.grid(True, linestyle="--", alpha=0.3)
plt.legend()

# Plot f'(x)
plt.subplot(1, 2, 2)
plt.plot(x, dy, label=r"$f'(x) = 2x$", color="orange")
plt.axhline(0, color="white", linewidth=0.5)
plt.axvline(0, color="white", linewidth=0.5)
plt.title("Derivative: $2x$")
plt.grid(True, linestyle="--", alpha=0.3)
plt.legend()

plt.tight_layout()
plt.show()
```

## Visual intuition

Imagine walking along the parabola.

At each step, you ask:

    “How steep is the ground right here?”

The derivative graph 2x answers that question.

- When you’re left of the vertex, the slope is negative → you’re walking downhill.
- At the vertex, slope is zero → flat.
- To the right, slope is positive → uphill.
- The farther you go, the steeper it gets → because 2x grows.

The straight line shape reflects the fact that the parabola’s slope increases at a constant rate.
