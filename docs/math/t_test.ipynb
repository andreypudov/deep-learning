# T-test

T-test is a statistical test used to determine if there is a significant difference between the means of two groups. It is commonly used in hypothesis testing to compare the means of two samples or to compare a sample mean to a known population mean.

$$
\begin{aligned}
t_k &= \frac{\bar{x} - \bar{y}}{s/\sqrt{n}}
&= \frac{\text{Difference in means}}{\text{Standard deviations}}
\end{aligned}
$$

Where:
- $\bar{x}$: mean of the first sample
- $\bar{y}$: mean of the second sample
- $s$: standard deviation of the samples
- $n$: number of observations in each sample

How to interpret $t$

Large $t$ → groups are far apart relative to their variability → stronger evidence of a real difference

Small $t$ → groups overlap a lot → weak evidence

## What the p‑value means

If the groups were actually identical, what is the probability of observing a difference this large (or larger) just by random chance?

Interpretation:
- $p$ < 0.05 → statistically significant (common threshold)
- $p$ < 0.01 → strong evidence
- $p$ < 0.001 → very strong evidence
- $p$ > 0.05 → not enough evidence to claim a difference

## How $t$ and $p$ relate

- Bigger $t$ → smaller $p$ → stronger evidence of a real difference
- Smaller $t$ → larger $p$ → weaker evidence of a real difference

## Mental model to remember

- $t$ tells you how big the difference is relative to noise
- $p$ tells you how surprising that difference is under the assumption of no effect
