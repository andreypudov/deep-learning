# Entropy

Entropy in information theory: everything is great and happy and, surprising things convey more information.

$$
H = -\sum_{i=1}^{n} p(x_i)\,\log_2\!\big(p(x_i)\big)
$$

Where $p(x_i)$ is the probability of $x_i$.

** Cross-entropy

Cross-entropy is a measure of the difference between two probability distributions.

$$
H(p, q) = -\sum_{i=1}^{n} p(x_i)\,\log_2\!\big(q(x_i)\big)
$$

Where $p$ is the true distribution and $q$ is the predicted distribution.

In deep learning, events are happen or do not happen -> $p(x)$ is either 0 or 1.
