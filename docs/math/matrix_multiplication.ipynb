# Matrix multiplication

$M{\times}N$ matrix - $M$ rows, $N$ columns (MR NiCe guy)

$$M{\times}N \cdot N{\times}K = M{\times}K$$

The number of columns in the first matrix must be equal to the number of rows in the second matrix.

The resulting matrix has the number of rows of the first matrix and the number of columns of the second matrix.


# Example

$$
\begin{pmatrix}0 & 1\\2 & 3\end{pmatrix}
{\times}
\begin{pmatrix}a & b\\c & d\end{pmatrix}
=
\begin{pmatrix}0{\cdot}a + 1{\cdot}c & 0{\cdot}b + 1{\cdot}d\\2{\cdot}a + 3{\cdot}c & 2{\cdot}b + 3{\cdot}d\end{pmatrix}
$$
