# Dot product

In mathematics, the dot product is an algebraic operation that takes two equal-length sequences of numbers (usually coordinate vectors), and returns a single number.

$$
\alpha = a \cdot b
       = \langle a, b \rangle
       = a^T b
       = \sum_{i=1}^{n} a_i b_i
       = a_1 b_1 + a_2 b_2 + \dots + a_n b_n
$$


## Example for vectors

$$
v=\begin{pmatrix}1\\0\\2\\5\\-2\end{pmatrix},
\qquad
w=\begin{pmatrix}2\\8\\-6\\1\\0\end{pmatrix}
$$

$$
\begin{aligned}
v^T w
&= 1\cdot 2 + 0\cdot 8 + 2\cdot(-6) + 5\cdot 1 + (-2)\cdot 0 \\
&= 2 + 0 - 12 + 5 + 0 \\
&= -5
\end{aligned}
$$

## Example for matrices:

$$
v=\begin{pmatrix}0 & 3 & 2\\-3 & -3 & 1\\1 & 0 & 2\end{pmatrix},
\qquad
w=\begin{pmatrix}1 & 0 & 6\\2 & -1 & 0\\5 & 1 & 4\end{pmatrix}
$$

$$
\begin{aligned}
v^T w
&= 0\cdot 1 + 3\cdot 0 + 2\cdot 6 + (-3)\cdot 2 + (-3)\cdot(-1) + 1\cdot 0 + 1\cdot 5 + 0\cdot 1 + 2\cdot 4 \\
&= 0 + 0 + 12 - 6 + 3 + 0 + 5 + 0 + 8 \\
&= 22
\end{aligned}
$$

## Interpretation of the dot product

A single number that reflects the commonalities between two objects (vectors, matrices, tensors, signals, images, etc.). The larger the dot product, the more similar the two objects are. The smaller the dot product, the less similar they are. If the dot product is zero, then the two objects are orthogonal (i.e., they have no commonalities).
