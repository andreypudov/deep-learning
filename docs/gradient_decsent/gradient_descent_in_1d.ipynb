# Gradient descent in 1D

$$
\begin{align*}
f(x) &= 3x^2 - 3x + 4 \\
f'(x) &= 6x - 3
\end{align*}

```python
import numpy as np
import matplotlib.pyplot as plt


def f(x):
    return 3 * (x**2) - 3 * x + 4


def df(x):
    return 6 * x - 3


x = np.linspace(-3, 3, 400)

y = f(x)
dy = df(x)

plt.figure(figsize=(8, 5))

plt.plot(x, y, label=r"$f(x) = -x^4 + 3x^2$", color="lightblue")
plt.plot(x, dy, label=r"$f'(x) = -4x^3 + 6x$", color="orange")

plt.axhline(0, color="white", linewidth=0.5)
plt.axvline(0, color="white", linewidth=0.5)

plt.xlim(-2, 2)
plt.ylim(-15, 20)

plt.title("Function and Derivative")
plt.grid(True, linestyle="--", alpha=0.3)
plt.legend()

plt.tight_layout()
plt.show()
```
