```python
# Exerciese: Pythagorean theorem
# c**2 = a**2 + b**2

# get two inputs from the user, a and b
# compute c using the Pythagorean theorem

a = float(input("Please enter the length of side a: "))
b = float(input("Please enter the length of side b: "))

c = (a**2 + b**2) ** 0.5
print("The length of the hypotenuse is:", c)
```
