```python
# Exercise: write a function that flips a coin N times, and reports the average

# import numpy as np


def flip_coins(times):
    probabilities = np.random.random(times)
    return probabilities


def average(probabilities):
    length = len(probabilities)
    summ = sum(probabilities)

    return summ / length


def flip_and_return_average(times):
    probabilities = flip_coins(times)

    return average(probabilities)


print("Flip coins and return the average: ", flip_and_return_average(10))
```

```python
import numpy as np

# np.mean(np.random.randn(100) > 0)


def coinFlip(N):
    propCoinFlips = np.mean(np.random.randn(N) > 0)
    print(str(N) + " coin flips had " + str(100 * propCoinFlips) + "% heads.")


coinFlip(20)
```
