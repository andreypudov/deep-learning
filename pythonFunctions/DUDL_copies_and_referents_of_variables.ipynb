```python
a = [4, 3]
b = a

b[0] = 5

print(a)
print(b)
```

```python
print(id(a))
print(id(b))
```

```python
a = [4, 3]

b = a[:]
b[0] = 5

print(a)
print(b)
```

```python
print(id(a))
print(id(b))
```

```python
a = {"q": 1, "w": 2}
b = a

print(id(a))
print(id(b))
```

```python
import copy

a = {"q": 1, "w": 2}
b = copy.deepcopy(a)

print(id(a))
print(id(b))
```
