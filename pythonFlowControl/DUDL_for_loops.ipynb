```python
# Exercise:
# Fibonacci series: [0, 1, 1, 2, 3, 5, 8, 13, ...]
# create a function that reports the Fib. series for input N


def fibonocci(n):
    if n <= 2:
        return 1

    return fibonocci(n - 1) + fibonocci(n - 2)


print(fibonocci(10))
```

```python
def fibonocci2(n):
    result = 0
    for i in range(1, n + 1):
        result += i

    return result


print(fibonocci2(10))
```

```python
def fibonocci3(n):
    v = [0, 1, 1]

    for i in range(n - 2):
        v.append(sum(v[-2:]))

    return v


print(fibonocci3(10))
```
