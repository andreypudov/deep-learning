```python
# Exercise: poisson counter

# 1) specify a target number np.exp(-lambda) | e^lambda -> 10
# 2) initialize a counter = 0, currval = 1
# 3) multiply currval by a random number U[0, 1]
# 4) repeat step 3 until currval < target
# 5) return the number of times that you go through steps 4

import numpy as np


def poissonCounter(lamb):
    counter = 0
    currval = 1
    target = np.exp(-lamb)

    while currval > target:
        currval = currval * np.random.rand()
        counter = counter + 1

    return counter


poissonCounter(10)
```
