```python
# Exercise: to compare three methods of creating an array
# N = 10000
# r[i] = i ** 2
# time test!

import time

N = 10_000_000

start = time.perf_counter()
python_array = [None] * N
for i in range(N):
    python_array = i**2
end = time.perf_counter()
print("Python array (preallocated): " + str(end - start))

start = time.perf_counter()
python_array = []
python_array.append(N)
for i in range(N):
    python_array = i**2
end = time.perf_counter()
print("Python array (list preallocated 2): " + str(end - start))

start = time.perf_counter()
python_array = []
for i in range(N):
    python_array.append(i**2)
end = time.perf_counter()
print("Python array (append): " + str(end - start))

start = time.perf_counter()
python_array = np.array([])
for i in range(N):
    np.append(python_array, i)
end = time.perf_counter()
print("NP array (np array): " + str(end - start))

start = time.perf_counter()
python_array = np.zeros(N)
for i in range(N):
    python_array[i] = i**2
end = time.perf_counter()
print("NP array (np zeroes): " + str(end - start))
```
