```python
import numpy as np

# if statements

# conditional
if True:
    print("This is true!")

if False:
    print("This is not printed")

print("Outside if")
```

```python
x = 4

if x == 4:
    print("The analysis finished correctly!")
```

```python
if "mike":
    print("Hi Mike!")

if "":
    print("This is not printed")
```

```python
r = np.random.randn()

if r > 0:
    print("The number is positive.")
if r < 0:
    print("The number is negative.")
```

```python
if r > 0:
    print("The number is positive.")
else:
    print("The number is negative.")
```

```python
if r > 0:
    outcome = "positive."
else:
    outcome = "negative."

print("The number is " + outcome)
```
