```python
def funfun(x, y):
    # check that inputs are numeric
    if not isinstance(x, (int, float)):
        if isinstance(x, str) and x.isnumeric():
            x = float(x)
            print("Note: converted x to float.")
        else:
            raise Exception("Input x numst be a number!")

    if not isinstance(y, (int, float)):
        if isinstance(y, str) and x.isnumeric():
            y = float(y)
            print("Note: converted y to float.")
        else:
            raise Exception("Input y numst be a number!")

    z = x * y
    return z


funfun("2", 3)
```
