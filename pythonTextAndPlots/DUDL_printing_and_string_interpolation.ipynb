```python
num = 11.7
sng = "cellar doors"

print("I would like to have " + str(num) + " " + sng + ".")
```

```python
print("I would like to have %g %s." % (num, sng))
```

```python
print("I would like to have %.3f %s." % (num, sng))
```

```python
num = 11.75384754
print("I would like to have %.3f %s." % (num, sng))
```

```python
# f-strings

print(f"I would like to have {num} {sng}.")
```

```python
print(f"I would like to have {num:.2f} {sng}.")
```

```python
print("I would like to have {n:.2f} {s}.".format(n=num, s=sng))
```

```python
# Exerciese
# print out each letter of the alphabet with its ordinal position
# a is the 1st letter
# c is the 3rd letter


def ordinaltg(n):
    return str(n) + {1: "st", 2: "nd", 3: "rd"}.get(
        4 if 10 <= n % 100 < 20 else n % 10, "th"
    )


for index, ch in enumerate(range(ord("a"), ord("z"))):
    print(f"{chr(ch)} is the {ordinaltg(index + 1)} letter")
```

```python
import string

letters = string.ascii_lowercase

for i in range(len(letters)):
    if i == 0 or i == 20:
        ordind = "st"
    elif i == 1 or i == 21:
        ordind = "nd"
    elif i == 2 or i == 22:
        ordind = "rd"
    else:
        ordind = "th"
    print(f"{letters[i]} is the {i + 1}{ordind} letter of the alphabet.")
```
