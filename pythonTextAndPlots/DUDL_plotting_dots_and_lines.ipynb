```python
import matplotlib.pyplot as plt

plt.plot(3, 4, "o")
plt.plot(3, 4.5, "bo", label="blue dot")
plt.legend()
plt.show()

plt.plot(4, 5, "x")
plt.plot(5, 6, "p")
plt.show()

plt.plot(3, 4, "r^", label="red triangle")
plt.legend(["blue", "red"])
# plt.legend()
```

```python
x = [0, 1]
y = [0, 2]

# p1 = (0, 0), p2 = (1, 2)
plt.plot(x, y, "m")
```

```python
x = [0, 1, 3]
y = [0, 2, -1]

plt.plot(x, y, "ms-")
plt.show()
```

```python
x = [0, 1, 3]
y = [0, 2, -1]

plt.plot(x, y, "ms:")
plt.show()
```

```python
import numpy as np

x = np.linspace(0, 3 * np.pi, 101)
y = np.sin(x)

plt.plot(x, y, "m:")
plt.show()
```
