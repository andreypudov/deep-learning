```python
def superDuperAwesomeFunction(input1, input2):
    prod = input1 * input2
    summ = input1 + input2
    return prod, summ


superDuperAwesomeFunction(10, -7)
```

```python
result = superDuperAwesomeFunction(10, -7)
prod = result[0]
prod
```

```python
prod, summ = superDuperAwesomeFunction(10, -7)
print(prod)
print(summ)
```

```python
# Exercise: create a function that computes a factorial
# 4! = 4 * 3 * 2 * 1
# compare against math.factorial()

import math


def factorial(value):
    if value == 1:
        return 1

    return value * factorial(value - 1)


print("my.factorial: ", factorial(4))
print("math/factorial: ", math.factorial(4))
```
