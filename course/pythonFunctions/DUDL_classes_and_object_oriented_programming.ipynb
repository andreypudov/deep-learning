```python
# OOP

# class: blueprint for creating a set of attributes and methods
# instance
```

```python
class model(object):
    # constructor method
    def __init__(self, numlayers, numunits, name):
        self.layers = numlayers  # attributes
        self.units = numunits
        self.name = name
        self.weights = 0

    def howManyUnits(self):
        totalUnits = self.layers * self.units
        print(f"There are {totalUnits} in the model.")

    def trainTheModel(self, x):
        self.weights += x
        return self.weights

    def __str__(self):
        return f"This is a {self.name} architecture."
```

```python
m1 = model(2, 5, "CNN")
m1
```

```python
m1.howManyUnits()
```

```python
m1 = model(12, 5, "CNN")
m1.howManyUnits()
```

```python
m1 = model(12, 5, "CNN")
m1.trainTheModel(2.5)
```

```python
m1.trainTheModel(2.5)
```

```python
# Exercise: create a new class
# 1) weights should be a random numbers matrix of size layer X units
# 2) change training method to multiply input X and add input Y

import numpy as np


class model2(object):
    # constructor method
    def __init__(self, numlayers, numunits, name):
        self.layers = numlayers  # attributes
        self.units = numunits
        self.name = name
        self.weights = np.random.randn(self.layers, self.units)

    def howManyUnits(self):
        totalUnits = self.layers * self.units
        print(f"There are {totalUnits} in the model.")

    def trainTheModel(self, x, y):
        self.weights = self.weights * x + y

    def __str__(self):
        return f"This is a {self.name} architecture with weights {self.weights}"


m = model2(2, 3, "RNN")
print(m)
m.trainTheModel(10, 1)
print(m)
```
