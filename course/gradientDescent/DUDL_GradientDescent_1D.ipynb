```python
# import all necessary modules
import numpy as np
import matplotlib.pyplot as plt

import matplotlib_inline.backend_inline

matplotlib_inline.backend_inline.set_matplotlib_formats("svg")
```

# Gradient descent in 1D

```python
# function (as a function)
def fx(x):
    return 3 * x**2 - 3 * x + 4


# derivative function
def deriv(x):
    return 6 * x - 3
```

```python
# plot the function and its derivative

# define a range for x
x = np.linspace(-2, 2, 2001)

# plotting
plt.plot(x, fx(x), x, deriv(x))
plt.xlim(x[[0, -1]])
plt.grid()
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend(["y", "dy"])
plt.show()
```

```python
# Gradient Descent

# random starting point
localmin = np.random.choice(x, 1)
print(localmin, type(localmin))

# learning parameters
learning_rate = 0.01
training_epochs = 100

# run through training
for i in range(training_epochs):
    grad = deriv(localmin)
    localmin = localmin - learning_rate * grad

localmin
```

```python
# plot the results

plt.plot(x, fx(x), x, deriv(x))
plt.plot(localmin, deriv(localmin), "ro")
plt.plot(localmin, fx(localmin), "ro")

plt.xlim(x[[0, -1]])
plt.grid()
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend(["f(x)", "df", "f(x) min"])
plt.title("Empirical local minimum: %s" % localmin[0])
plt.show()
```

# Store the model parameters and outputs on each iteration

```python
# random starting point
localmin = np.random.choice(x, 1)

# learning parameters
learning_rate = 0.01
training_epochs = 1000

# run through training and store all the results
modelparams = np.zeros((training_epochs, 2))
for i in range(training_epochs):
    grad = deriv(localmin)
    localmin = localmin - learning_rate * grad
    modelparams[i, 0] = localmin[0]
    modelparams[i, 1] = grad[0]
```

```python
# plot the gradient over iterations

fig, ax = plt.subplots(1, 2, figsize=(12, 4))

for i in range(2):
    ax[i].plot(modelparams[:, i], "o-")
    ax[i].set_xlabel("Iteration")
    ax[i].set_title(f"Final estimated minimum: {localmin[0]:.5f}")

ax[0].set_ylabel("Local minimum")
ax[1].set_ylabel("Derivative")

plt.show()
```

```python
# Most often in DL, the model trains for a set number of iterations, which is what we do here. But there are other ways
#    of defining how long the training lasts. Modify the code so that training ends when the derivative is smaller than
#    some threshold, e.g., 0.1. Make sure your code is robust for negative derivatives.

# random starting point
localmin = np.random.choice(x, 1)
print(localmin)

# learning parameters
learning_rate = 0.01
training_epochs = 100
training_threshold = 0.1

number_of_iterations = 0

# run through training
for i in range(training_epochs):
    grad = deriv(localmin)
    localmin = localmin - learning_rate * grad

    if np.absolute(grad) < training_threshold:
        break

    number_of_iterations += 1

print("Number of training iterations: ", number_of_iterations)

# plot the results

plt.plot(x, fx(x), x, deriv(x))
plt.plot(localmin, deriv(localmin), "ro")
plt.plot(localmin, fx(localmin), "ro")

plt.xlim(x[[0, -1]])
plt.grid()
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend(["f(x)", "df", "f(x) min"])
plt.title("Empirical local minimum: %s" % localmin[0])
plt.show()
```

# Analysis

This does not generate a more accurate result with the threshold that we were given (0.1). It also does not generate a more accurate result in general, because the purpose of gradient descent is to converge to a global (that's the goal) minima, or at least some kind of minima. Stopping when the derivative gets really small just prevents it from getting further precision, but still produces a decent result.


## Potential Problems

As elaborated on previously, a potential problem is that it will prevent the algorithm from obtaining a more precise answer - and will yield a "good enough" answer. Thus, you will always get close to the right answer but will never quite reach it this approach, and so all learning rate/training epoch combinations that converge to the minima will never get a more precise empirical minima estimation because a predefined threshold will prevent this from occurring. In fact, they will all yield the same answer, provided that your initial choice of a local minima is the same.

```python
# normalized gradient

# random starting point
localmin = np.random.choice(x, 1)
print(localmin)

# learning parameters
learning_rate = 0.01
training_epochs = 100
step_length = 0.1

number_of_iterations = 0

# run through training
for i in range(training_epochs):
    grad = deriv(localmin)
    # localmin = localmin - learning_rate * grad

    direction = grad / np.linalg.norm(grad)
    localmin = localmin - step_length * direction

    if np.absolute(grad) < training_threshold:
        break

    number_of_iterations += 1

print("Number of training iterations: ", number_of_iterations)

# plot the results

plt.plot(x, fx(x), x, deriv(x))
plt.plot(localmin, deriv(localmin), "ro")
plt.plot(localmin, fx(localmin), "ro")

plt.xlim(x[[0, -1]])
plt.grid()
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend(["f(x)", "df", "f(x) min"])
plt.title("Empirical local minimum: %s" % localmin[0])
plt.show()
```


🏁 Bottom line

You can use constant‑length steps, but:
- it overshoots in steep regions
- it oscillates near minima
- it converges slowly or not at all
- it ignores valuable information in the gradient magnitude
