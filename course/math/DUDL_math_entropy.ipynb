# Reminder of entropy:

$$H(p) = -\sum_x p(x)\log(p(x))$$

```python
# probability of an event happening
p = 0.25

# NOT the correct formula!
H = -(p * np.log(p))
print("Wrong entropy: " + str(H))
```

```python
# probability of an event happening
p = 0.25

# probability of an event not happening
q = 1 - p

# we need to sum oll values of x
```

```python
# the correct way to compute entropy
x = [0.25, 0.75]

H = 0
for p in x:
    H -= p * np.log(p)

print("Correct entropy: " + str(H))
```

```python
# also correct, written out for N=2 events
# Binary cross-entropy
H = -(p * np.log(p) + (1 - p) * np.log(1 - p))
print("Correct entropy: " + str(H))
```

# Cross-entropy

```python
# note: all probs must sum to 1!
p = [1, 0]  # sum=1
q = [0.25, 0.75]  # sum=1

H = 0
for i in range(len(p)):
    H -= p[i] * np.log(q[i])

print("Cross entropy: " + str(H))
```

```python
# also correct, written out for N=2 events
H = -(p[0] * np.log(q[0]) + p[1] * np.log(q[1]))
print("Correct entropy: " + str(H))

# simplification (in [1, 0], the second one is zero)
H = -np.log(q[0])
print("Manually simplified: " + str(H))
```

```python
# now using pytorch
import torch
import torch.nn.functional as F

# note: inputs must be Tensors
q_tensor = torch.Tensor(q)
p_tensor = torch.Tensor(p)

# pytorch doesnt work well with numpy
# F.binary_cross_entropy(p, q)

# pay attention on order of P and Q (the output should be the same as above)
print(F.binary_cross_entropy(p_tensor, q_tensor))
print(F.binary_cross_entropy(q_tensor, p_tensor))
```
