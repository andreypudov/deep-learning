```python
# Exercise
# Create a function that computes the "dot product" between two vectors
# [1, 2, 3]
# [0, 4, 1]
# 1 * 0 + 2 * 4 + 1 * 3 = 8 + 3 = 11
# 1) check the both inputs are numpy arrays (isinstance())
# 2) check that the twp inputs are the same length
# 3a) break out of the function and give a helpful error message (raise Exception())
# 3b) compute and return the dot product

import numpy as np


def dotProduct(input1: np.ndarray, input2: np.ndarray):
    if not (isinstance(input1, np.ndarray) and isinstance(input2, np.ndarray)):
        raise Exception(
            f"The type of provided vectors is not a numpy array: {type(input1)} {type(input2)}"
        )

    if len(input1) != len(input2):
        raise Exception(
            f"The length of provided vectors is not the same: {len(input1)} != {len(input2)}"
        )

    return sum(input1 * input2)


a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
c = np.array([10])

print(dotProduct(a, b))
# print(dotProduct(a, [4, 5, 6]))
# print(dotProduct(a, c))
```
