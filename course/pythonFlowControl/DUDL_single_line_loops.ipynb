```python
newlist = [""] * 10

for i in range(10):
    if i % 2 == 1:
        newlist[i] = "Odd"
    else:
        newlist[i] = i

newlist
```

```python
["Odd" if i % 2 == 1 else i for i in range(10)]
```

```python
import numpy as np

# repeat without using numpy
x1 = np.linspace(-3, 3, 11)
y1 = x1**2

print(x1)
print(y1)
```

```python
x1 = [i * 6 / 10 for i in range(-5, 6)]
y1 = [i**2 for i in x1]

print(x1)
print(y1)
```
