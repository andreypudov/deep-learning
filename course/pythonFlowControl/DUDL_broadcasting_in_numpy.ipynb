```python
import numpy as np

X = np.array([[1, 2, 3], [2, 3, 4], [3, 4, 5], [4, 5, 6]])
w = np.array([10, 20, 30])

print(X), print(" ")
print(w)
```

```python
X.shape
```

```python
X.shape[0]
```

```python
Y = np.zeros(X.shape)

for i in range(X.shape[0]):
    Y[i, :] = X[i, :] + w

Y
```

```python
# broadcasting

Z = X + w

print(Z), print(" ")
print(Y)
```

```python
Z = X - w

print(Z), print(" ")
print(Y)
```

```python
Z = X * w

print(Z), print(" ")
print(Y)
```

```python
Z = X / w

print(Z), print(" ")
print(Y)
```

```python
# Exercise:
# 1) create a vector of integers 0:8 (v)
# 2) reshape v into 3x3 matrix (M) (np.reshape)
# 3) repeat M into a 9x3 (C) (np.tile)
# 4) broadcast-multiple C with v (B)

v = np.array(range(0, 9))
M = np.reshape(v, (3, 3))
C = np.tile(M, (3, 1))
B = C * np.reshape(v, (len(v), 1))

B
```
