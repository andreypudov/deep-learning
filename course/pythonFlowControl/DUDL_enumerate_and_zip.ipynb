```python
# enumerate and zip

import numpy as np

ns = np.linspace(-5, 5, 7)

for i in range(len(ns)):
    print("Index " + str(i) + " has a value of " + str(ns[i]))
```

```python
for i, n in enumerate(ns):
    print("Index " + str(i) + " has a value of " + str(n))
```

```python
text = "Hello my name is Andrey."

vowels = np.zeros(len(text))

for index, character in enumerate(text):
    if character in "aeiou":
        vowels[index] = 1
```

```python
listA = [3, 4, 5, 6, 3, -17]
listB = ["q", "w", "e", "r", "t", "y"]

for i in range(len(listA)):
    print(str(listA[i]) + " " + listB[i])
```

```python
for a, b in zip(listA, listB):
    print(str(a) + " " + b)
```

```python
# Exercise
# given the following lists, create a dictionary using zip

names = ["alpha", "beta", "gamma"]
values = [10, 20, 30]

dictA = {}

for key, value in zip(names, values):
    dictA[key] = value

print(dictA)
```

```python
# dictB = dict()
dictB = dict(zip(names, values))

print(dictB)
```
