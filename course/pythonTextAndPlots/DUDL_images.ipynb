```python
from imageio import imread
import matplotlib.pyplot as plt
import numpy as np
```

```python
m = 3
n = 5

M = np.random.randint(10, size=(m, n))
print(M)

# plt.plot(M);
plt.imshow(M)

for i in range(m):
    for j in range(n):
        plt.text(
            j,
            i,
            str(M[i, j]),
            fontsize=16,
            horizontalalignment="center",
            verticalalignment="center",
        )

plt.colorbar()
plt.show()
```

```python
# Exercise: Visualize the Hibert matrix using seaborn (heatmap)
# Hibert matrix: Hij = 1 / (i + j - 1)
# H is 10x10

import seaborn as sns

rows = 10
columns = 10

H = np.zeros((rows, columns))

for row in range(rows):
    for column in range(columns):
        H[row, column] = 1 / (row + column + 1)

sns.heatmap(H, vmax=0.4)
plt.show()
```
